### Generative AI Application with OpenAI Model

```python
from langchain_groq import ChatGroq

model = ChatGroq(
    model="llama-3.1-8b-instant",
    temperature=0.6
)

# model.invoke("Hello, how are you?")
```

```python
response=model.invoke("Write me an essay on AI")
print(response)
```

```python
print(response.content)
```

```python
from langchain_google_genai import GoogleGenerativeAI
model=GoogleGenerativeAI(model="gemini-2.5-flash-lite")
response=model.invoke("Explain me about Artificial Intelligence")
print(response)
```

### GROQ Model Integration using init

```python
import os
from langchain.chat_models import init_chat_model

init_chat_model("groq:qwen/qwen3-32b")
response=model.invoke("write me a joke on Tollywood film heroes in telugu")
print(response)
```

### Streaming and Batch

```python
for chunk in model.stream("write me a joke on Tollywood film heroes in telugu"):
    print(chunk, end="",flush=True)
```

# Batch

## Batching a collection of independent requests to a model can significantly improve performance and reduce costs, as the processing  can be done in parallel.

```python
response =  model.batch(
    [
        "What is the capital of France?",
        "What is the capital of Germany?",
        "What is the capital of Italy?",
    ]
)

for response in response:
    print(response)
```
